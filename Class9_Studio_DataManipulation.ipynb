Import the CSVs for each of the following cities: Baltimore, Boston, New York, and Philadelphia. Set up a dataframe for each city.

```python
# Import the necessary libraries and CSVs. Make some dataframes!

import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np

df_baltimore = pd.read_csv("dataset/baltimore_9-24-2016_9-30-2017.csv")
df_boston = pd.read_csv("dataset/boston_9-24-2016_9-30-2017.csv")
df_newyork = pd.read_csv("dataset/new-york_9-24-2016_9-30-2017.csv")
df_philadelphia = pd.read_csv("dataset/philadelphia_9-24-2016_9-30-2017.csv")
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!
combined_cities = df_baltimore.append([df_boston, df_newyork, df_philadelphia])

for col in combined_cities.columns:
    pct_missing = np.mean(combined_cities[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
for col in combined_cities.columns:
    pct_missing = np.mean(combined_cities[col].isnull())
    if(pct_missing==1):
        combined_cities=combined_cities.drop([col], axis=1)

for col in combined_cities.columns:
    pct_missing = np.mean(combined_cities[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

## Answer Some Questions

Use `groupby()` and `agg()` to answer the following two questions:

1. What is the mean low and high prices for each type of unit of sale in the Northeast region? In the last lesson, we learned that a unit of sale could be something like a bin or individually. 
2. What is the average number of pumpkins for each variety that came into terminal markets for the year by region? Pumpkin varieties include Howden and Fairytale pumpkins.

```python
# Put your code here to find the mean low and high prices in the Northeast region for each type of unit of sale.

mean_per_unit = combined_cities.groupby("Unit of Sale")[["Low Price", "High Price"]].agg("mean")
mean_per_unit
```

```python
# Put your code here to find the average number of pumpkins coming into terminal markets of each variety.

average_pumpkins_per_variety = combined_cities.groupby("City Name")
average_pumpkins_per_variety["Variety"].value_counts()
```

```python
# Import three cities

import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np

df_chicago = pd.read_csv("dataset/chicago_9-24-2016_9-30-2017.csv")
df_detroit = pd.read_csv("dataset/detroit_9-24-2016_9-30-2017.csv")
df_stlouis = pd.read_csv("dataset/st-louis_9-24-2016_9-30-2017.csv")
```

```python
# Combine the Midwest cities into one dataset. Print percentage of missing values. 

midwest_cities = df_chicago.append([df_detroit, df_stlouis])

for col in midwest_cities.columns:
    pct_missing = np.mean(midwest_cities[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Clearn unnessecary data (= remove 100% missing values)

for col in midwest_cities.columns:
    pct_missing = np.mean(midwest_cities[col].isnull())
    if(pct_missing ==1):
        midwest_cities = midwest_cities.drop( [col], axis=1)

for col in midwest_cities.columns:
    pct_missing = np.mean(midwest_cities[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```
